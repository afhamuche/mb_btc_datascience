```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# read in the data
data = pd.read_csv("bigtable_trades-20230320-2.csv")
```

```python
# create a pivot table to aggregate the data by bool_type, bool_volume, and var_mean
table = pd.pivot_table(data, values='timestamp', index=['bool_type'], columns=['bool_volume'], aggfunc=lambda x: len(x.unique()), fill_value=0)
```

```python
# add labels and title
plt.title('Relationship between bool_type, bool_volume, and var_mean')
plt.xlabel('bool_volume')
plt.ylabel('bool_type')
```

```python
# create the heatmap using seaborn
sns.heatmap(table, annot=True, fmt='g', cmap='viridis')
```

```python
# display the plot
plt.show()
```
